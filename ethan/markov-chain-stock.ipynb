```python
import numpy as np
import plotly.express as px
```

```python
c = [[0.3,0.5,1.0], #up breakpoints
    [0.4,0.6,1.0], #same breakpoints
    [0.4,0.7,1.0]] #down breakpoints

meanings = {
    0: 'up',
    1: 'same',
    2: 'down'
}

trials = 10000
i = np.random.randint(0,2) #0 is up, 1 is same, 2 is down
X_t = [meanings[i]] #markov chain
```

```python
for t in range(trials):
    u = np.random.rand()
    j = 0
    while c[i][j] <= u:
        j += 1
    i = j
    X_t.append(meanings[i])
```

```python
fig = px.histogram(X_t)
fig.show()
```
