```python
import numpy as np
import plotly.express as px
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'up': [],
    'same': [],
    'down': []
}
df = pd.DataFrame.from_dict(data,orient='columns')
```

```python
c = [[0.3,0.5,1.0], #up breakpoints
    [0.4,0.7,1.0], #same breakpoints
    [0.3,0.7,1.0]] #down breakpoints

meanings = [
    'up',
    'same',
    'down'
]

trials = 10000
i = np.random.randint(0,2) #0 is up, 1 is same, 2 is down
p = np.zeros(3)
p[i] += 1
df.loc[len(df)] = p
```

```python
for t in range(trials):
    u = np.random.rand()
    j = 0
    while c[i][j] <= u:
        j += 1
    i = j
    p = np.zeros(3)
    p[i] += 1
    df.loc[len(df)] = p
```

```python
axis_0 = df.sum(axis=0)
print(axis_0)
plt.figure(dpi=100)
plt.bar(axis_0.index,(axis_0)/len(df))
plt.show()
```

```python
cumsum = df.cumsum(axis=0)
for i in range(len(cumsum)):
    cumsum.loc[i] = cumsum.loc[i]/(i+1)
```

```python
plt.figure(dpi=100)
for c in cumsum.columns:
    plt.plot(cumsum[c],label=c)
plt.legend()
plt.show()
```
