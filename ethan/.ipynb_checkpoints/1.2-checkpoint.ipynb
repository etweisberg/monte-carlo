```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
red = np.random.randint(1,6,1000)
green = np.random.randint(1,6,1000)
dice = red + green
```

```python
x=np.arange(0, 13, 0.1).tolist()
```

```python
y = []
for val in x:
    y.append(len(dice[dice<=val])/len(dice))
plt.plot(x,y)
plt.xlabel('Dice Value')
plt.ylabel('cdf of Dice Value')
```
