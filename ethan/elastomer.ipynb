```python
# Testing out P(R) distribution

def gen_polymer():
    end_to_ends = []

    for i in range(1000):
        n_steps = 1000
        path = np.random.randint(0,2,n_steps)
        path[path==0] = -1
        square_end_to_end = np.sum(path)**2
        end_to_ends.append(square_end_to_end)
    return np.average(end_to_ends)
```

```python
R_values = []
for i in range(1000):
    R_values.append(math.sqrt(gen_polymer()))
```

```python
plt.title('Normal Distribution vs. Simulated Distribution of R')
plt.hist(R_values,density=True,bins=20)

mu = np.average(R_values)
sigma = np.std(R_values)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma),linewidth=4)
plt.show()
```
