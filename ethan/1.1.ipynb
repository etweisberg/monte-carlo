# Buffon's Needle

```python
import numpy as np
import math
import plotly.express as px
```

```python
throws = 10000000
```

```python
x=np.random.rand(throws) # random (0,1]
```

```python
theta=np.random.rand(throws)*0.5*math.pi # random (0,pi/2]
```

```python
hits = []
for i in range(throws):
    hits.append(1 if x[i] <= 0.5 * math.sin(theta[i]) else 0) #is a hit if x <= L/2*sin(theta)
```

```python
throws/np.sum(hits)
```

```python
pi_estimates = []
for i in range(10000):
    throws = 10000
    x=np.random.rand(throws) # random (0,1]
    theta=np.random.rand(throws)*0.5*math.pi # random (0,pi/2]
    hits = []
    for i in range(throws):
        hits.append(1 if x[i] <= 0.5 * math.sin(theta[i]) else 0) #is a hit if x <= L/2*sin(theta)
    pi_estimates.append(throws/np.sum(hits))
fig = px.histogram(pi_estimates)
fig.show()
```
