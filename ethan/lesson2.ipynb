```python
import random
import matplotlib.pyplot as plt
```

```python
def play(total_money, bet_money, total_plays):
    
    n_plays = []
    money = []
    
    choice = random.randint(1,2)
    if choice == 1:
        special = 11
    if choice == 2:
        special = 10
    
    play = 1
    
    for play in range(total_plays):
        draw = random.randint(1,100)
        if draw % 2 == choice % 2 and draw != special:
            total_money += bet_money
            n_plays.append(play)
            money.append(total_money)
        else:
            total_money -= bet_money
            n_plays.append(play)
            money.append(total_money)
    
    plt.ylabel('Player Money in $')
    plt.xlabel ('Number of bets')
    plt.plot(n_plays,money)
    final_money=money[-1]
    return (final_money)
```

```python
final_money = []
for i in range(10):
    final_money.append(play(10000,100,50))
print(final_money)
print(sum(final_money))
print(f'The players net change was {sum(final_money)/len(final_money)-10000} $')
```

```python
profits = []
for i in range(100):
    profits.append(play(10000,100,50)-10000)
```

```python
import numpy as np
import scipy.stats

def mean_confidence_interval(data, confidence):
    a = 1.0 * np.array(data)
    n = len(a)
    m, se = np.mean(a), scipy.stats.sem(a)
    h = se * sc.stats.t.ppf((1 + confidence) / 2., n-1)
    return ("Mean: {}".format(m) + "\n" + "Lower end of {}% CI: {}".format(confidence*100,m-h) + "\n" + "Upper end of {}% CI: {}".format(confidence*100,m+h))

print(mean_confidence_interval(profits,0.95))
```
